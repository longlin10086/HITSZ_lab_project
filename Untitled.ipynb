```python
import serial
import matplotlib.pyplot as plt
ser = serial.Serial('COM9', 9600, timeout=1)


plt.ion()

x_data = []
y_data = []


fig, ax = plt.subplots()
x = 0
while True:
    x += 1
    if data := ser.readline().decode('utf-8').rstrip():
        y = data
        x_data.append(float(x))
        y_data.append(float(y))
        ax.plot(x_data, y_data, 'b-')
        fig.canvas.draw()
        plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 创建Figure对象和Axes对象
fig, ax = plt.subplots()

# 绘制曲线
ax.plot(x, y)

# 设置图像标题和坐标轴标签
ax.set_title('Sin Wave')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# 显示图像
plt.show()
```

```python
import serial
import matplotlib.pyplot as plt

# 打开串口，设置波特率和超时时间
ser = serial.Serial('COM9', 9600, timeout=1)

# 创建Figure对象和Axes对象
fig, ax = plt.subplots()

# 设置曲线图样式和坐标轴标签
ax.set_title('Sensor Data')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Value')

# 打开交互模式，并绘制空曲线图
plt.ion()
line, = plt.plot([], [], color='blue')

# 初始化x轴和y轴数据
x_data = []
y_data = []

# 连续读取数据并绘制曲线图
while True:
    # 读取一行数据，并将其转换为浮点数
    data = ser.readline().decode().strip()
    if data:
        value = float(data)
        
        # 将数据添加到x轴和y轴数据列表中
        x_data.append(len(x_data))
        y_data.append(value)
        
        # 更新曲线图数据
        line.set_data(x_data, y_data)
        ax.relim()
        ax.autoscale_view()
        
        # 刷新曲线图
        plt.draw()
        plt.pause(0.01)
        
plt.ioff()
plt.show()
```
