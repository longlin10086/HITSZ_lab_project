```python
from matplotlib import pyplot
kp = 0.01
ki = 0.95
kd = 0.001

err_last_last = 0
err_last = 0
err_now = 0
target = 80

period = list(range(1, 500, 20))
v = []
x = 0

for i in period:

    d_u = kp*(err_now-err_last)+ki*err_now+kd*(err_now+err_last_last-2*err_last)
    x += d_u
    if x > 255:
        x = 255
    if x < -255:
        x = -255
    v.append(x)
    
    err_last_last = err_last
    err_last = err_now
    err_now = target - x
    
pyplot.plot(period, v)
```
